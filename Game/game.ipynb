```python
class Player1(Player):
    
    def update(self):       

        self.rect.x += self.change_x
        self.rect.y += self.change_y

    def go_up(self):
        self.change_y = -chy    

    def go_down(self):
        self.change_y = chy
        
        
class Player2(Player):
    
    def update(self):

        self.calc_grav()

        self.rect.x += self.change_x

        block_hit_list = pygame.sprite.spritecollide(self, self.level.platform_list, False)
        for block in block_hit_list:
            if self.change_x > 0:
                self.rect.right = block.rect.left
            elif self.change_x < 0:
                self.rect.left = block.rect.right
            
        self.rect.y += self.change_y    
        
        block_hit_list = pygame.sprite.spritecollide(self, self.level.platform_list, False)
        for block in block_hit_list:
            if self.change_y > 0:
                self.rect.bottom = block.rect.top
            elif self.change_y < 0:
                self.rect.top = block.rect.bottom
            self.change_y = 0
    
    def calc_grav(self):
        if self.change_y == 0:
            self.change_y = gravs
        else:
            self.change_y += gravm
            
        if self.rect.y >= SCREEN_HEIGHT - self.rect.height and self.change_y >= 0:
            self.change_y = 0
            self.rect.y = SCREEN_HEIGHT - self.rect.height

    def jump(self):
        self.rect.y += jh
        platform_hit_list = pygame.sprite.spritecollide(self, self.level.platform_list, False)
        self.rect.y -= jh
        if len(platform_hit_list) > 0 or self.rect.bottom >= SCREEN_HEIGHT:
            self.change_y = jump_height
```

```python
class Move(pygame.sprite.Sprite):
    def __init__(self, image, x, y, speed_x, speed_y):
        super().__init__()
        self.image = pygame.image.load(image)
        self.rect = self.image.get_rect(center = (x, y))
        self.speed_x = speed_x
        self.speed_y = speed_y
        
    def update(self):
        self.rect.move_ip(self.speed_x, self.speed_y)
        if self.rect.right < 0:
            self.kill()
```

```python
# платформы
class Platform(pygame.sprite.Sprite):
    def __init__(self, width, height):
        super().__init__()

        self.image = pygame.image.load(platform)

        self.rect = self.image.get_rect()

class Level(object):
    def __init__(self, player):
        self.platform_list = pygame.sprite.Group()
        self.player = player

    def update(self):
        self.platform_list.update()

    def draw(self, screen): 
        screen.blit(bg, (0, 0))
        self.platform_list.draw(screen)

class Level_01(Level):
    def __init__(self, player):
        Level.__init__(self, player)
        
        #координаты платформ
        level = [[210, 32, 150, 550],
                [210, 32, 300, 450],
                [210, 32, 510, 320],
                [210, 32, 130, 320],
                [210, 32, 300, 200],
                [210, 32, 100, 80]]
        
        for platform in level:
            block = Platform(platform[0], platform[1])
            block.rect.x = platform[2]
            block.rect.y = platform[3]
            block.player = self.player
            self.platform_list.add(block)
```

```python
#выводим текст на экран
def draw_text(surf, text, size, x, y):
    font = pygame.font.Font(font_name, size)
    text_surface = font.render(text, True, WHITE)
    text_rect = text_surface.get_rect()
    text_rect.midtop = (x, y)
    surf.blit(text_surface, text_rect)
```

```python
class end_of_game(pygame.sprite.Sprite):
    def __init__(self, bg, text, pic, screen, x, y, end):
        super().__init__()
        self.bg = bg
        self.text = text
        self.pic = pic
        self.screen = screen
        self.x = x
        self. y = y
        self.end = end
    def func(self):
        self.screen.blit(self.bg,(0,0))
        draw_text(self.screen, self.text, 64, self.x, self.y)
        pygame.display.flip()
        while True:
            for event in pygame.event.get():
                if event.type == pygame.QUIT:
                    return -1
                if event.type == pygame.KEYDOWN:
                    if event.key == pygame.K_SPACE:
                        if self.pic == self.end:
                            return self.pic
                        self.screen.blit(history[self.pic-1], (0,0))
                        pygame.display.flip()   
                        self. pic+=1
```

```python
# Основная функция программы

def main():
    
    pygame.init()
    
    # Установка высоты и ширины
    screen = pygame.display.set_mode(size)
    
    finish = end_of_game(bg, "You lose", 9, screen, SCREEN_WIDTH / 2, SCREEN_HEIGHT / 4, 9)

    # Название игры
    pygame.display.set_caption(Name)
    
    player1 = Player1(1)
    player2 = Player2(2)

    #декорации
    ADDENEMY = pygame.USEREVENT + 1
    pygame.time.set_timer(ADDENEMY, 350)

    ADDCLOUD = pygame.USEREVENT + 2
    pygame.time.set_timer(ADDCLOUD, 1200)

    ADDCACTUS = pygame.USEREVENT + 3
    pygame.time.set_timer(ADDCACTUS, 1500)
    
    level_list = []
    level_list.append(Level_01(player2))

    current_level = level_list[0]
    
    enemies = pygame.sprite.Group()
    cactuses = pygame.sprite.Group()
    player_1 = pygame.sprite.Group()
    player_2 = pygame.sprite.Group()
    
    player_1.add(player1)
    player_2.add(player2)
    
    player2.level = current_level
    
    player1.rect.x = startx1
    player1.rect.y = starty1
    player2.rect.x = startx2
    player2.rect.y = starty2
    
    # Цикл будет до тех пор, пока пользователь не нажмет кнопку закрытия
    # FPS
    clock = pygame.time.Clock()
    
    #переменные для отслеживания этапа игры
    done=False
    game_over_1 = game_over_2 = False
    game0 = True
    game1 = game2 = False
    pic = 1
    score_1 = score_2 = 0

    # Основной цикл программы
    while not done:
        # Отслеживание действий
        for event in pygame.event.get():
            if event.type == pygame.QUIT: # Если закрыл программу, то останавливаем цикл
                done = True
            if game0:
                first = end_of_game(bg, "Press SPACE", pic, screen, SCREEN_WIDTH / 2, SCREEN_HEIGHT / 4, 4)
                pic = end_of_game.func(first)
                if pic == -1:
                    done = True
                game0 = False
                game1 = True
            if game_over_1: 
                second = end_of_game(bg, "You win! Press SPACE", pic, screen, SCREEN_WIDTH / 2, SCREEN_HEIGHT / 4, 6)
                pic = end_of_game.func(second)
                if pic == -1:
                    done = True
                game1 = False
                game2 = True
                game_over_1 = False
            if game_over_2: 
                third = end_of_game(bg, "You win! Press SPACE", pic, screen, SCREEN_WIDTH / 2, SCREEN_HEIGHT / 4, 9)
                pic = end_of_game.func(third)
                game2 = False
                game_over_2 = False
                done = True
                
            if event.type == ADDENEMY:
                new_enemy = Move(enemy, random.randint(SCREEN_WIDTH + 20, SCREEN_WIDTH + 100),
                random.randint(0, SCREEN_HEIGHT), random.randint(-20, -5), random.randint(-7, 7))
                enemies.add(new_enemy)
                player_1.add(new_enemy)
                
            if event.type == ADDCLOUD:
                new_cloud = Move(cloud_pic, random.randint(SCREEN_WIDTH + 20, SCREEN_WIDTH + 100),
                random.randint(0, SCREEN_HEIGHT), -5, 0)
                player_1.add(new_cloud)
                player_2.add(new_cloud)
                
            if event.type == ADDCACTUS:
                new_cactus = Move(cactus[random.randint(0, 5)], random.randint(SCREEN_WIDTH + 20, SCREEN_WIDTH + 100), 
                random.randint(0, SCREEN_HEIGHT), random.randint(-15, -5), random.randint(-7, 7))
                cactuses.add(new_cactus)
                player_2.add(new_cactus)
                
            if game1:
                if pygame.sprite.spritecollide(player1, enemies, False):
                    player1.kill()
                    end_of_game.func(finish)
                    done = True

                if event.type == pygame.KEYDOWN:
                    
                    if event.key == pygame.K_LEFT or event.key == pygame.K_a:
                        player1.go_left()
                    if event.key == pygame.K_RIGHT or event.key == pygame.K_d:
                        player1.go_right()
                    if event.key == pygame.K_UP or event.key == pygame.K_w:
                        player1.go_up()
                    if event.key == pygame.K_DOWN or event.key == pygame.K_s:
                        player1.go_down()
                        
                if event.type == pygame.KEYUP:
                    
                    if event.key == pygame.K_LEFT or event.key == pygame.K_a and player.change_x < 0:
                        player1.stop_x()
                    if event.key == pygame.K_RIGHT or event.key == pygame.K_d and player.change_x > 0:
                        player1.stop_x()
                    if event.key == pygame.K_UP or event.key == pygame.K_w and player.change_y < 0:
                        player1.stop_y()
                    if event.key == pygame.K_DOWN or event.key == pygame.K_s and player.change_y > 0:
                        player1.stop_y()
                        
                
                if player1.rect.right > SCREEN_WIDTH:
                    player1.rect.right = SCREEN_WIDTH
                if player1.rect.left < 0:
                    player1.rect.left = 0
                if player1.rect.y < 0:
                    player1.rect.y = 0        
                if player1.rect.y >= SCREEN_HEIGHT - player1.rect.height and player1.change_y >= 0:
                    player1.stop_y()
                    player1.rect.y = SCREEN_HEIGHT - player1.rect.height 
                    
                if not game_over_1:
                    score_1 +=1
                if score_1 == final_1:
                    game_over_1 = True
                    
            if game2:
                if event.type == pygame.KEYDOWN:
                    if event.key == pygame.K_LEFT or event.key == pygame.K_a:
                        player2.go_left()
                    if event.key == pygame.K_RIGHT or event.key == pygame.K_d:
                        player2.go_right()
                    if event.key == pygame.K_UP or event.key == pygame.K_SPACE or event.key == pygame.K_w:
                        player2.jump()

                if event.type == pygame.KEYUP:
                    if event.key == pygame.K_LEFT or event.key == pygame.K_a and player2.change_x < 0:
                        player2.stop_x()
                    if event.key == pygame.K_RIGHT or event.key == pygame.K_d and player2.change_x > 0:
                        player2.stop_x()
                        
                if player2.rect.right > SCREEN_WIDTH:
                    player2.rect.right = SCREEN_WIDTH
                if player2.rect.left < 0:
                    player2.rect.left = 0
                if player2.rect.y < 0:
                    player2.rect.y = 0  
                if player2.rect.y >= SCREEN_HEIGHT - player2.rect.height and player2.change_y >= 0:
                    player2.kill()
                    end_of_game.func(finish)
                    done = True

                # Проверяем столкновение с млнстрами и убиваем игрока
                if pygame.sprite.spritecollide(player2, cactuses, False):
                    player2.kill()
                    end_of_game.func(finish)
                    done = True
                    
                if not game_over_2:
                    score_2 +=1
                if score_2 == final_2:
                    game_over_2 = True   

        screen.blit(bg,(0,0))
        
        if game1: 
            draw_text(screen, str(score_1), 18, SCREEN_WIDTH / 2, 10)
            player_1.update()
            player_1.draw(screen)
            
        if game2: 
            draw_text(screen, str(score_2), 18, SCREEN_WIDTH / 2, 10)
            current_level.update()
            current_level.draw(screen)
            player_2.update()
            player_2.draw(screen)

        clock.tick(FPS)

        pygame.display.flip()       

    pygame.quit()
```
